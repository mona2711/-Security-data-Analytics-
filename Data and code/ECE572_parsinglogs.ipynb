```python
from datetime import datetime
```

```python
file = open('snort','r')
# file = open('test_file.txt','r')
output_file = open('parsed_file.csv','a')
# output_file = open('test_file.csv','a')

output_file.write('Action Performed,Connection Classification,Priority,Date-Time,Source IP,Source Port,Destination IP,Destination Port,Protocol,TimeToLive,TypeOfService,ID,IP Length,DataGram Length, Frag Offset, Frag Size, Flag -2,Flag-1, Flag-U, Flag-A, Flag-P, Flag-R, Flag-S, Flag-F,All Flags,Seq, Ack, Win, TCP length, TCP Options, NOP NOP TS, MSS, NOP WS, SackOK TS\n')

for i in file.readlines():
#     line - 1
    if '[**]' in i:
        i = i.replace(',',' ')
        output_file.write(i[i.find(']',5)+1:-5]+' ,')        
        
#     line - 2
    if 'Priority' in i:
        if 'Classification' in i:
            output_file.write(i[17:i.find(']',17)]+' ,')        
            output_file.write(i[i.find('Priority')+10:i.find('Priority')+11]+' ,')            
        else:
            output_file.write(' ,')
            output_file.write(i[i.find('Priority')+10:i.find('Priority')+11]+' ,')
            
#   line - 3
    if '->' in i:
        params_1 = i.split(' ')
        for j in params_1:
            if '\n' in j:
                j=j[:-1]
            if not(j=='->'):             
                if not(j[2]=='/'):
                    ip_port = j.split(':')
                    if len(ip_port)==2:
                        j = ip_port[0]+','+ip_port[1]
                    else:
                        j = ip_port[0]+','
                else:
                    j = '2019/'+j 
                    seconds = datetime.strptime(j,  '%Y/%m/%d-%H:%M:%S.%f')
                    j = seconds.timestamp()
                output_file.write(str(j)+' ,')                
                
#     line - 4
    if ('[**]' not in i) and ('Priority' not in i) and ('->' not in i) and ('=>' not in i) and ('***' not in i) and ('Frag Offset' not in i) and ('Type:0' not in i):
        params_2 = i.split(':')
        
        if not(i=='\n'):                    
            output_file.write(i.split(' ')[0]+',')
            
        if ' Len: ' in i:
            params_2 = params_2[:-1]

        params_2 = params_2[1:]   
        for j in params_2:
            if '\n' in j:
                j = j[:-1]
            output_file.write(j.split(' ')[0]+' ,')
            
#     line - 5

    if 'Frag Offset' in i:
        k = i.split(' ')
        k[7]= k[7].replace('\n',' ')
        temp = k[2]+','+k[7]
        output_file.write(temp)

    if '***' in i:
        output_file.write(', ,')

        k = i.split(':')
        for l in k:
            if '***' in l:
                count = 0 
                for flag in l:
                    count = count+1
                    output_file.write(flag+',')
                    if count == 8:                        
                        break
                
            if '\n' in l:
                l=l[:-1]
            val = l.split(' ')
            if not(val[0]==''):
                output_file.write(val[0]+' ,')
                
            else:
                output_file.write(val[1]+' ,')
                
#     line - 6
    if 'TCP Options' in i:
        output_file.write(i[i.find('TCP Options')+13:i.find('TCP Options')+14]+' ,')
        
        q = i.split('=>')[1]
        if 'NOP NOP TS' in q:       
            p = q[q.find(' NOP NOP TS')+13:].split(' ')
            output_file.write(p[0]+' '+p[1]+' ,')
            
        else:
            output_file.write(' ,') 
            
        if 'MSS' in q:
            p = q[q.find(' MSS')+6:].split(' ')
            output_file.write(p[0]+' ,')
            
        else:
            output_file.write(' ,')
            
        if 'NOP WS' in q:
            p = q[q.find('NOP WS')+8:].split(' ')
            output_file.write(p[0]+' ,')
            
        else:
            output_file.write(' ,')
            
        if 'SackOK TS' in q:
            p = q[q.find('SackOK TS')+11:].split(' ')
            output_file.write(p[0]+' '+p[1]+' ,')                            
            
        else:
            output_file.write(' ,')   
        
    if i=='\n':
        output_file.write('\n')
        
```

```python
output_file.close()
```
